```python
from hapsira.core.iod import vallado
from hapsira.core.iod import izzo
from astropy import units as u
import numpy as np
from hapsira.bodies import Earth

k = Earth.k.to(u.km ** 3 / u.s ** 2)
r1 = np.array([15945.34, 0.0, 0.0]) * u.km
r2 = np.array([12214.83899, 10249.46731, 0]) * u.km
tof = 4560 * u.s # Time of flight
v1izzo, v2izzo = izzo(k, r1.value, r2.value, tof.value, M=0, prograde=True, lowpath=True, numiter=35, rtol=1e-8) # M, prograde, lowpath, numiter, rtol are all default values from Poliastro
v1izzo = v1izzo * u.km / u.s
v2izzo = v2izzo * u.km / u.s
print(v1, v2)
```
